### Simple histogram

```python
times = Time(obs.frbcat['utc'])
```

```python
times.jyear
```

```python
bins = np.arange(2001, 2020,1)
bins
```

```python
plt.clf()
ax = plt.gca()
ax.hist(times.jyear, bins=bins)
#
plt.show()
```
